```python
def summarize_one_split(experiment_name,
                        pred_name,
                        criteria,
                        acc_col):
    res_dir = Path(f"../results/{experiment_name}")
    data = dict()
    for tsv_path in sorted(res_dir.rglob(f"{pred_name}/evaluation_results/{criteria}_summary.tsv")):
        df = pd.read_csv(tsv_path, sep='\t')
        model_name = tsv_path.parent.parent.parent.stem
        model_name = model_name.split("_")[-1]
        data["Index"] = df['Unnamed: 0'].values
        data["num_examples"] = df['num_examples'].values
        data[model_name] = df[acc_col].values
        
    result = pd.DataFrame.from_dict(data)
    return result
```

```python
result = summarize_one_split("mist_cf_nist_fast_filter_sweep_k",
                        "preds_COMMON_fast_filter",
                        "mass_acc",
                        "Top 1 acc.")
result
```
